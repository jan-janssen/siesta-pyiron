Based on https://wiki.fysik.dtu.dk/ase/ase/calculators/siesta.html

```python
import os
from ase import Atoms
from ase.calculators.siesta import Siesta
from ase.units import Ry
```

```python
a0 = 5.43
bulk = Atoms(
    'Si2', 
    [(0, 0, 0), (0.25, 0.25, 0.25)],
    pbc=True
)
b = a0 / 2
bulk.set_cell(
    [(0, b, b), (b, 0, b), (b, b, 0)], 
    scale_atoms=True
)
```

```python
calc = Siesta(
    label='Si',
    xc='PBE',
    mesh_cutoff=200 * Ry,
    energy_shift=0.01 * Ry,
    basis_set='DZ',
    kpts=[10, 10, 10],
    fdf_arguments={
        'DM.MixingWeight': 0.1,
        'MaxSCFIterations': 100
    },
    pseudo_path=os.path.abspath('.'),
    pseudo_qualifier='',
    directory='si_run'
)
bulk.set_calculator(calc)
```
